```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("adult.data.csv", header=0)
```

```python
df
```

```python
# 1. Race distribution
print("Race distribution:")
print(df['race'].value_counts(), "\n")
```

```python
# 2. Average age of men
avg_age_men = round(df[df['sex'] == "Male"]['age'].mean(), 1)
print(f"Average age of men: {avg_age_men}")
```

```python

# 3. Percentage with Bachelors degrees
percent_bachelors = round((df[df['education'] == "Bachelors"].shape[0] / df.shape[0]) * 100, 1)
print(f"Percentage with Bachelors degrees: {percent_bachelors}%")
```

```python
# 4. Percentage with advanced education earning >50K
filters = ['Bachelors', 'Masters', 'Doctorate']
higher_edu = df[df['education'].isin(filters)]
percent_higher_edu_rich = round(higher_edu['salary'].value_counts(normalize=True)[">50K"] * 100, 1)
print(f"Percentage with higher education earning >50K: {percent_higher_edu_rich}%")
```

```python

# 5. Percentage without advanced education earning >50K
lower_edu = df[~df['education'].isin(filters)]
percent_lower_edu_rich = round(lower_edu['salary'].value_counts(normalize=True)[">50K"] * 100, 1)
print(f"Percentage without higher education earning >50K: {percent_lower_edu_rich}%")
```

```python
# 6. Min hours per week
min_hours = df['hours-per-week'].min()
print(f"Minimum hours per week: {min_hours}")
```

```python
# 7. Rich percentage among those who work minimum hours
min_workers = df[df['hours-per-week'] == min_hours]
rich_min_workers = round(min_workers['salary'].value_counts(normalize=True).get(">50K", 0) * 100, 1)
print(f"Percentage of rich among those who work fewest hours: {rich_min_workers}%")
```

```python

# 8. Country with highest percentage of rich (>50K)
rich_counts = df[df.salary == ">50K"].groupby("native-country")["salary"].count()
total_counts = df["native-country"].value_counts()
rich_percent = (rich_counts / total_counts) * 100
highest_rich_country = rich_percent.idxmax()
highest_rich_value = round(rich_percent.max(), 1)
print(f"Country with highest percentage of rich: {highest_rich_country}")
print(f"Highest percentage of rich people in country: {highest_rich_value}%")
```

```python

# 9. Most popular occupation for those earning >50K in India
top_IN_occupation = df[(df['native-country'] == "India") & (df['salary'] == ">50K")]['occupation'].value_counts().idxmax()
print(f"Top occupation for those earning >50K in India: {top_IN_occupation}")
```
